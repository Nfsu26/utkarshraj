```python
text = " ".join(df)
tokenizer = nltk.tokenize.TreebankWordTokenizer()
tokens = tokenizer.tokenize(text)
```

```python
sw = set(nltk.corpus.stopwords.words("english"))

lemitizer = nltk.WordNetLemmatizer()

tokens = [lemitizer.lemmatize(t) for t in tokens if t not in sw and t not in string.punctuation]
```

```python
vocab = set(tokens)
print("vocab : " , len(vocab))
print("token : " , len(tokens))
```

```python
comment_text = " ".join(tokens)
sw.add("br")
wc = WordCloud(width=700,height=700,background_color="Black",min_font_size=10,stopwords=sw).generate(comment_text)

plt.figure(figsize=(7,7),facecolor=None)
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
